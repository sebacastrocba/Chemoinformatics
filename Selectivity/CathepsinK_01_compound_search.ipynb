## BÃºsqueda de compuestos que inhiben Catepsina K

```python
import pandas as pd
import numpy as np
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
target_query = target.search('cathepsin')
targets = pd.DataFrame.from_dict(target_query)
targets.head()
```

```python
selected_target = targets[(targets['pref_name'] == 'Cathepsin K') & (targets['organism'] == 'Homo sapiens') ]
selected_target
```

```python
selected_target = selected_target.loc[0, 'target_chembl_id']
```

```python
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type= 'IC50')
```

```python
df = pd.DataFrame.from_dict(res)
df.head()
```

```python
df.info()
```

```python
df.standard_type.unique()
```

```python
df.standard_units.unique()
```

```python
df.isnull().sum()
```

```python
print(len(df))
df.duplicated(subset='molecule_chembl_id', keep='first').sum()
```

```python
df.drop_duplicates(subset=['molecule_chembl_id'], inplace=True)
```

```python
len(df)
```

```python
df2 = df[df.standard_value.notna()]
```

```python
len(df2)
```

```python
df2.reset_index(drop=True)
```

```python
df2.info()
```

```python
seleccion = ['molecule_chembl_id', 'canonical_smiles', 'standard_value', 'standard_units']
df3 = df2[seleccion]
df3.isnull().sum()
```

```python
df3.info()
```

```python
df3['standard_value'] = df3['standard_value'].apply(pd.to_numeric)
```

```python
df3 = df3.reset_index(drop=True)
df3
```

```python
df3.standard_units.unique()
```

```python
seleccion = df3[df3.standard_units == 'ug.mL-1']
```

```python
seleccion
```

```python
df3.to_csv('CathepsinK_01_compound_search.csv', index=False)
```

ConclusiÃ³n: Existen 1515 compuestos que inhibien catepsina K.

```python
bioactivity_class = []
for i in df3.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive_CTSK")
  elif float(i) <= 1000:
    bioactivity_class.append("active_CTSK")
  else:
    bioactivity_class.append("intermediate_CTSK")
```

```python
len(bioactivity_class)
```

```python
seleccion = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df4 = df3[seleccion]
df4 = df4.reset_index(drop=True)
```

```python
df5 = pd.concat([df3, pd.Series(bioactivity_class)], axis=1)
```

```python
df5.reset_index(drop=True, inplace=True)
```

```python
df5.drop('standard_units', axis=1, inplace=True)
```

```python
df6 = df5.rename(columns={0:'class'})
df6
```

```python
df6.to_csv('CTSK_bioactivity_preprocessed_data.csv', index=False)
```
